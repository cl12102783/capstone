## Dataset

There are 7 different data files:

* __application_train/application_test__: the main training and testing data with information about each loan application at Home Credit. Every loan has its own row and is identified by the `SK_ID_CURR`. The training application data comes with the `TARGET` with indicating 0: the loan was repaid and 1: the loan was not repaid. 
* __bureau__: data concerning client's previous credits from other financial institutions. Each previous credit has its own row in bureau and is identified by the `SK_ID_BUREAU`, Each loan in the application data (`SK_ID_CURR`) can have multiple previous credits.
* __bureau_balance__: monthly data about the previous credits in bureau. Each row is one month balance of a previous credit, and a single previous credit (`SK_ID_BUREAU`) can have multiple rows, one for each month balance of the credit length. 
* __previous_application__: previous applications for loans at Home Credit of clients who have loans in the application data. Each current loan in the application data (`SK_ID_CURR`) can have multiple previous loans. Each previous application has one row and is identified by the feature `SK_ID_PREV`. 
* __POS_CASH_BALANCE__: monthly data about previous point of sale or cash loans clients have had with Home Credit. Each row is one month of a previous point of sale or cash loan, and a single previous loan (`SK_ID_PREV`) can have many rows.
* __installments_payment__: payment history for previous loans at Home Credit. There is one row for every made payment and one row for every missed payment. 
* __credit_card_balance__: monthly data about previous credit cards clients have had with Home Credit. Each row is one month of a credit card balance, and a single previous loan (`SK_ID_PREV`) can have many rows.

The diagram below (provided by Home Credit) shows how the tables are related.

#### Encoding categorical variables:
In this project, label encoding (LabelEncoder in Scikit-Learn) is used for any binary categorical variables, and one-hot encoding (pandas get_dummies) is used for others.

```python
# Create a label encoding object
le = LabelEncoder()
le_count = 0

# Iterate through columns
for col in train:
    if train[col].dtype == 'object':
        # If 2 or fewer unique categories
        if len(list(train[col].unique())) <= 2:
            # Train on the training data
            le.fit(train[col])
            # Transform both training and testing data
            train[col] = le.transform(train[col])
            test[col] = le.transform(test[col])
            
            # Keep track of how many columns were label encoded
            le_count += 1
            
print('%d columns were label encoded.' % le_count)
```

```python
# One-hot encoding of categorical variables
train = pd.get_dummies(train)
test = pd.get_dummies(test)

print('Train data shape:',train.shape)
print('Test data shape:',test.shape)
```

There are a number of features with the extreme value of *365243* related to **DAYS**, also in other dataframes. Let's replace them with NAN.

```python
import re

def replace_day_outliers(df):
    """Replace 365243 with np.nan in any columns with DAYS"""
    for col in df.columns:
        if "DAYS" in col:
            df[col] = df[col].replace({365243: np.nan})

    return df

# Replace all the day outliers
train = replace_day_outliers(train)
test = replace_day_outliers(test)
bu = replace_day_outliers(bu)
bu_blc = replace_day_outliers(bu_blc)
cred_blc = replace_day_outliers(cred_blc)
pos_blc = replace_day_outliers(pos_blc)
pr_app = replace_day_outliers(pr_app)
ins_pay = replace_day_outliers(ins_pay)
```

## 2. Feature Selection
Now we have too many features to handle modeling and there are still more features to be added by feature engineeing later. So let's do feature selection first for all the features we have. The actions are:

* Removed any columns with greater than 75% missing values in the train or test set
* Removed collinear features as measured by the correlation coefficient greater than 0.9
* Keep features with 95% cumulative feature importances in Light GBM

In the end of this section, the number of features is **reduced from 1639 to 363**. More details of the code can be viewed in the original code.

### 2.1 Missing Values
An important consideration is the missing values in the dataframe. Columns with too many missing values might have to be dropped. Here we define 75% as the missing percentage threshold.

### 2.2 Collinear Variables

Collinearity can cause high variance in the prediction. Here, collinearity > 0.9 is defined as the threshold to sift features.

```python
# Track columns to remove and columns already examined
cols_to_remove = []
cols_seen = []
cols_to_remove_pair = []

# Iterate through columns and correlated columns
for key, value in above_threshold_vars.items():
    # Keep track of columns already examined
    cols_seen.append(key)
    for x in value:
        if x == key:
            next
        else:
            # Only want to remove one in a pair
            if x not in cols_seen:
                cols_to_remove.append(x)
                cols_to_remove_pair.append(key)
            
cols_to_remove = list(set(cols_to_remove))
print('Number of columns to remove: ', len(cols_to_remove))
```

We can remove these columns from both the training and the testing datasets. By using Light GBM as the testbed, the deletion of these features is validated to be with better performance.

```python
train = train.drop(columns = cols_to_remove)
test = test.drop(columns = cols_to_remove)

print('Training Corrs Removed Shape: ', train.shape)
print('Testing Corrs Removed Shape: ', test.shape)
```

### 3.1 Polynomial Features
In this method, we make features that are powers of existing features as well as interaction terms between existing features. While two variables by themselves may not have a strong influence on the target, combining them together into a single interaction variable might show a relationship with the target. *PolynomialFeatures* from *Scikit-Learn* is a useful class to realize this.

By considering the time consumption, only combination with degree of 2 (15 new added features) is considered here for **EXT_SOURCE** series and **DAYS_BIRTH**, which are analyzed as the most important features from section 1 and 2. ### 3.3 Time processing for installment payment data frame

The installments_payments data contains information on each payment made on the previous loans at Home Credit. It has two date offset columns:

* **DAYS_INSTALMENT**: number of days before current application at Home Credit that previous installment was supposed to be paid
* **DAYS_ENTRY_PAYMENT**: number of days before current application at Home Credit that previous installment was actually paid

Thus build a flag column to indicate whether the loan is late paid. Also, other feautures on the mutual relations are built.

```python
ins_pay['LATE_PAYMENT'] = ((pd.to_timedelta(ins_pay['DAYS_ENTRY_PAYMENT'], 'D') - 
                            pd.to_timedelta(ins_pay['DAYS_INSTALMENT'], 'D')).astype(int)
                           > 0).astype(int).astype('object')
```

```python
ins_pay['PAYMENT_PERC'] = ins_pay['AMT_PAYMENT'] / ins_pay['AMT_INSTALMENT']
ins_pay['PAYMENT_DIFF'] = ins_pay['AMT_INSTALMENT'] - ins_pay['AMT_PAYMENT']
# Days past due and days before due (no negative values)
ins_pay['DPD'] = ins_pay['DAYS_ENTRY_PAYMENT'] - ins_pay['DAYS_INSTALMENT']
ins_pay['DBD'] = ins_pay['DAYS_INSTALMENT'] - ins_pay['DAYS_ENTRY_PAYMENT']
ins_pay['DPD'] = ins_pay['DPD'].apply(lambda x: x if x > 0 else 0)
ins_pay['DBD'] = ins_pay['DBD'].apply(lambda x: x if x > 0 else 0)
```
